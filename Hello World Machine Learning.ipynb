```python
# Se crean las listas con las variables a utilizar
var_disc = ["weekday_is_monday", "weekday_is_tuesday", "weekday_is_wednesday", "weekday_is_thursday", 
            "weekday_is_friday", "weekday_is_saturday", "weekday_is_sunday", "is_weekend", 
            "data_channel_is_lifestyle", "data_channel_is_entertainment", "data_channel_is_bus",
            "data_channel_is_socmed", "data_channel_is_tech", "data_channel_is_world"]
var_cont = ["timedelta", "n_tokens_title", "n_tokens_content", "n_unique_tokens", "n_non_stop_words",
            "n_non_stop_unique_tokens", "num_hrefs", "num_self_hrefs", "num_imgs", "num_videos", 
            "average_token_length", "num_keywords", "data_channel_is_lifestyle", 
            "data_channel_is_entertainment", "data_channel_is_bus", "data_channel_is_socmed", 
            "data_channel_is_tech", "data_channel_is_world", "kw_min_min", "kw_max_min", "kw_avg_min", 
            "kw_min_max", "kw_max_max", "kw_avg_max", "kw_min_avg", "kw_max_avg", "kw_avg_avg", 
            "self_reference_min_shares", "self_reference_max_shares", "self_reference_avg_sharess", 
            "LDA_00", "LDA_01", "LDA_02", "LDA_03", "LDA_04", "global_subjectivity", 
            "global_sentiment_polarity", "global_rate_positive_words", "global_rate_negative_words", 
            "rate_positive_words", "rate_negative_words", "avg_positive_polarity", "min_positive_polarity",
            "max_positive_polarity", "avg_negative_polarity", "min_negative_polarity", 
            "max_negative_polarity", "title_subjectivity", "title_sentiment_polarity", 
            "abs_title_subjectivity", "abs_title_sentiment_polarity"]
target = "shares"
```

```python
# Quita los espacios a los extremos en los nombres de las columnas
df.columns = [x.strip() for x in df.columns]
```

```python
# Hace un recorte del dataframe, donde solamente se trae las variables seleccionadas
aux = df[var_cont+var_disc + [target]]
```

## 3. Modelado

### 3.1. Estandarización de Datos

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
Xs = pd.DataFrame(data = sc.fit_transform(X = df[var_cont]), columns = var_cont)
```

### 3.2. Componentes Principales

```python
from sklearn.decomposition import PCA
cp = PCA()
Xc = pd.DataFrame(data = cp.fit_transform(X = df[var_cont]), columns = var_cont)
```

### 3.3. Modelación no supervisada

#### 3.3.1. Análisis Cluster

```python
from sklearn.cluster import KMeans
km = KMeans(n_jobs=-1, n_clusters=3)
df["cl"] = km.fit_predict(X=Xs)
```

```python
df["cl"].value_counts(normalize = True)
```

```python
df[var_cont + ["cl"]].groupby("cl").agg(["min", "max", "mean"])
```

### 3.4. Modelación Supervisada

#### 3.4.1. Modelos de regresión

```python
# Se crean las muestras de entrenamiento y validación
from sklearn.model_selection import cross_val_score, train_test_split
X = aux[var_cont + var_disc]
y = aux[[target]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

##### 3.4.1.1. Regresión Lineal

```python
from sklearn.linear_model import LinearRegression
# Inicializa el modelo
reglin = LinearRegression(n_jobs=-1)
# Entrena el modelo
reglin.fit(X = X_train, y = y_train)
```

```python
# Calcula la métrica de ajuste de la muestra de entrenamiento
print reglin.score(X = X_train, y = y_train)
# Calcula la métrica de ajuste de la muestra de validación
print reglin.score(X = X_test, y = y_test)
```

```python
from sklearn.metrics import r2_score, make_scorer
```

```python
# Calida que el modelo generalice, es decir, que no sobreajuste
ls_r = cross_val_score(cv=10, estimator=reglin, n_jobs=-1, scoring=make_scorer(r2_score), X = X, y = y)
print np.mean(ls_r), ls_r
```

```python
# Muestra la importancia de las variables
zip(var_cont + var_disc, reglin.coef_[0])
```

#### 3.4.2. Modelos de clasificación

```python
# Se crea la variable objetivo
aux["y_disc"] = (aux[target] > np.median(aux[target]))*1
```

```python
# Se crean las muestras de entrenamiento y validación
from sklearn.model_selection import cross_val_score, train_test_split
X = aux[var_cont + var_disc]
y = aux["y_disc"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

##### 3.4.2.1. Regresión Logística

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(n_jobs=-1)
logreg.fit(X = X_train, y = y_train)
```

```python
# Calcula la métrica de ajuste de la muestra de entrenamiento
print logreg.score(X = X_train, y = y_train)
# Calcula la métrica de ajuste de la muestra de validación
print logreg.score(X = X_test, y = y_test)
```

```python
# Valida que el modelo generalice, es decir, que no sobreajuste
ls_r = cross_val_score(cv=10, estimator=logreg, n_jobs=-1, scoring="roc_auc", X = X, y = y)
print np.mean(ls_r), ls_r
```

#### 3.4.3. Árbol de decisión

```python
from sklearn.tree import DecisionTreeClassifier
arbol = DecisionTreeClassifier()
arbol.fit(X = X_train, y =y_train)
```

```python
# Calcula la métrica de ajuste de la muestra de entrenamiento
print arbol.score(X = X_train, y = y_train)
# Calcula la métrica de ajuste de la muestra de validación
print arbol.score(X = X_test, y = y_test)
```

```python
# Valida que el modelo generalice, es decir, que no sobreajuste
ls_r = cross_val_score(cv=10, estimator=arbol, n_jobs=-1, scoring="roc_auc", X = X, y = y)
print np.mean(ls_r), ls_r
```

### 3.5. Selección de mejor modelo

```python
aux["y_disc_pred"] = logreg.predict(X)
```
